```scala
Erika Bailon




Search




Channel Sidebar

Jump to...



Threads
ChannelsChannels

4308-project
general
links
random
shirtdesigns

Add a channel

Direct MessagesDirect Messages

Slackbot

Chakrya Ros(you)

Ali

dennisnguyen

Erika Bailon

Jason Lubrano

Javier

sky


Invite people

AppsApps

Karma

Polly


Add apps

Untitled 
// HW 3
// Maxwell Wenzel, Erika Bailon, Michael Gigiolio
import util.Random 
// Main function that runs through the days for the simulation 
object Main extends App {
    var theShop = new Store()
    for(day <- 1 to 35){
        theShop.dailyRentals()
        if(day == 35){ // On last day print report instead of incrementing the day
            theShop.endReport()
        }else{
            theShop.incrementDay()
        }
    }
}
/* the master store class that the simulation interfaces with */
class Store(){
    private var revenue = 0
    // Map of all the customers
    private val customers = Map(1 -> new Casual(1), 2 -> new Casual(2), 3 -> new Casual(3),
                                                4 -> new Casual(4), 5 -> new Regular(5), 6 -> new Regular(6),
                                                7 -> new Regular(7), 8 -> new Regular(8), 9 -> new Business(9),
                                                10 -> new Business(10))
    // Map of all the items
    private val items =  Map("brush" -> new Item("brush", "pa"), "roller" -> new Item("roller", "pa"), 
                            "bucket" -> new Item("bucket", "pa"), "tray" -> new Item("tray", "pa"),
                            "rake" -> new Item("rake", "ya"), "mower" -> new Item("mower", "ya"),
                            "hoe" -> new Item("hoe", "ya"), "shovel" -> new Item("shovel", "ya"),
                            "mixer" -> new Item("mixer","co"), "wheelbarrow" -> new Item("wheelbarrow","co"),
                            "pump" -> new Item("pump", "co"), "scraper" -> new Item("scraper", "co"),
                            "saw" -> new Item("saw", "wo"),"level" -> new Item("level", "wo"),
                            "sander" -> new Item("sander", "wo"), "drill" -> new Item("drill", "wo"),
                            "wrench" -> new Item("wrench", "pl"), "snake" -> new Item("snake", "pl"),
                            "cutter" -> new Item("cutter", "pl"), "caulk" -> new Item("caulk", "pl")) 
    // List of available items
    private var availItems: List[String] = items.keys.toList
    // List of all the rentals
    private var rentals: List[Rental] = List()
    
    // Function that creates a rental for a customer
    private def makeRental(custID: Int, numItems: Int, numNights: Int) = {
        var tRented: List[String] = availItems.take(numItems)
        availItems = availItems.drop(numItems) // get items available for rental
        var cost = 0
        for(it <- tRented){ // calc cost
            cost += items(it).price
        }
        val totCost = cost * numNights // calc total cost
        revenue += totCost
        rentals = new Rental(custID, tRented, numNights, totCost) +: rentals // Add rental to list
    }
    // Day incrementer used directly in simulation
    def incrementDay() = {
        for(rent <- rentals){
            if(rent.increment()){// call increment function and return items as rentals are up
                availItems = availItems ++ rent.toolsRented
                customers(rent.custID).returnTools(rent.toolsRented.length)
            }
        }
    }
    // Get the customers that came into store and call the make rental
    def dailyRentals() = {
        var randCust: List[Int] = List()
        // get random customers that can rent
        for(id <- 1 to 10){
            if(customers(id).canRent()){
                randCust = randCust :+ id
            }
        }
        if(randCust.length != 0){ // Check if anyone can come
            // get a random selection of the customers that come in
            randCust = Random.shuffle(randCust).take(1 + Random.nextInt(randCust.length - 1))
            var order = (0,0)
            
            for(cust <- randCust){
                order = customers(cust).getRental(availItems.length)
                if(order != (0,0)){ // check if each person available came
                    makeRental(cust, order._1, order._2) // make their rental
                }
            }
        }
    }
    // Printer fucntion for the end of the simulation
    def endReport(){
        val toolsAvail = availItems.length
        println(s"$toolsAvail Available Tools")
        for(t <- availItems){
            println(t)
        }
        println(s"Total Money Made: $revenue")
        println("ACTIVE RENTALS")
        for(r <- rentals){
            if(!r.isPast()){
                r.report()
            }
        }
        println("PAST RENTALS")
        for(r <- rentals){
            if(r.isPast()){
                r.report()
            }
        }
    }
}
/* Rentals*/
// stores information for each rental made
class Rental(val custID: Int, val toolsRented: List[String] ,private val length:Int, private val cost:Int){
    private var daysLeft = length // number of days remaining in rental
    def report() = { // printing function for end report
        println(s"Customer $custID rented:")
        for(ite <- toolsRented){
            println(ite)
        }
        print(s"For $length day")
        if(length > 1){
            print("s")
        }
        println(s" costing $cost \n")
    }
    def increment(): Boolean = { // decreases the number of days left
        daysLeft = daysLeft - 1
        if(daysLeft == 0){// returns true if items are to be returned
            true
        }else{
            false
        }
    }
    def isPast():Boolean = { // Checks if rental is active or not for printing
        if(daysLeft > 0){
            false
        } else {
            true
        }
    }
}
/* Customer Stuff */
// abstract class for customers with common behavior between types
abstract class Customer(val id:Int){
    protected var numTools:Int = 0
    def canRent():Boolean = { // Check if a customer has already their max rentals
        if(numTools == 3){
            false
        } else {
            true
        }
    }
    
    def getRental(curTools:Int): (Int, Int)
    def returnTools(numReturn:Int) = {
        numTools = numTools - numReturn 
    }
}
// different subclasses which implement the different rental behaviors
// return a tuple with the number of items rented and how for how long as ints
// return (0,0) if they there aren't enough tools left
class Casual(val i:Int) extends Customer(i) {
    override def getRental(curTools:Int): (Int, Int) = {
        var request = 1
        if(numTools < 2){
            request = request + Random.nextInt(2)
        }
        if(request > curTools){
            (0, 0)
        } else {
            numTools = numTools + request
            (request, 1 + Random.nextInt(2))
        }
    }
}
class Regular(val i:Int) extends Customer(i) {
    override def getRental(curTools:Int): (Int, Int) = {
        var request = 1 + Random.nextInt(3 - numTools)
        if(request > curTools){
            (0, 0)
        } else {
            numTools = numTools + request
            (request, 3 + Random.nextInt(3))
        }
    }
}
class Business(val i :Int) extends Customer(i) {
    override def getRental(curTools:Int): (Int, Int) = {
        if(3 > curTools){
            (0, 0)
        } else {
            numTools = 3
            (3, 7)
        }
    }
}
// item class used to determine cost of each type
class Item(val name: String, val typ: String){
    val price:Int = typ match {
        case "pa" => 5
        case "ya" => 15
        case "co" => 20
        case "wo" => 7
        case "pl" => 10
    }
}






```
